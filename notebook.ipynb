```python
import os 

PATH = r'.venv'
Nfiles = 0
Ndirs = 0
for root, dirnames, filenames in os.walk(PATH):
    Ndirs += len(dirnames)
    Nfiles += len(filenames)
    
print("Ndirs", Ndirs)
print("Nfiles", Nfiles)
```

```python
# Autre m√©thode pour raccourcir les imports et importer seulement une fonction d'une librairie
from os import walk as oswalk

PATH = r'.venv'

Nfiles = 0
Ndirs = 0

for root, dirnames, filenames in oswalk(PATH):
    Ndirs += len(dirnames)
    Nfiles += len(filenames)
    
print("Ndirs", Ndirs)
print("Nfiles", Nfiles)
```

```python
debut = 0
fin = 9
nbpoint = 10
t = np.linspace(debut, fin, nbpoint)
t
```

```python
fig = plt.figure(figsize=(2,2))

a = 1
b = 0
c = 1.2
d = -.1
funct = a*t + b
funct2 = c*t + d

plt.plot(t, funct, "-.", color="black", lw=3, label="f1")
plt.plot(t, funct2, ":", color="red", lw=3, label="f2")
plt.grid()
plt.xlim((t.min(), t.max()))
plt.ylim((np.minimum(funct.min(),funct2.min()), np.maximum(funct.max(),funct2.max())))
plt.legend()
```
