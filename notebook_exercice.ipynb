9. Créez une fonction qui prend un tableau NumPy de nombres entiers en entrée et renvoie un nouveau tableau contenant uniquement les nombres premiers du tableau d'origine.

```python
# 2️⃣ Fonction pour tester si un nombre est premier
def isPrime(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True    

def est_premier(tab):
    premier = []
    for i in tab :
        if isPrime(i) :
            premier.append(int(i))
    return premier

f = np.random.randint(20,size=(10)) 
print(f)
print(est_premier(f))
```
