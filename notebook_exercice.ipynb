```python

tab = [5, 4, 6, 6, 3, 3, 4]

#?(i=0, i=len(tab), i=i+1)

#! len(tab) = longueur du tableau
#! i est l'index
#! i = 0 démarer à la première valeur du tableau

#! for i in range (i = index)
#! for element in tab = element = valeur des élements du tableau
#! i et element sont juste des conventions tu peux très bien l'appeler caca ou x 

def sorting(liste):
    for i in range(len(liste)):
        mini = i
        for j in range(i+1, len(liste)):
            if liste[j] < liste[mini]:
                mini = j
        temp = liste[i]
        liste[i] = liste[mini]
        liste[mini] = temp
    return liste

def ordre_croissant(tab):
    tab2 = []                # nouvelle liste triée
    while len(tab) > 0:      # tant qu'il reste des éléments
        mini = tab[0]        # on suppose que le premier est le plus petit
        for element in tab:  # on cherche le vrai minimum 
            if element < mini:
                mini = element
        tab2.append(mini)    # on ajoute le minimum à la nouvelle liste
        tab.remove(mini)     # on l'enlève de l'ancienne liste
    return tab2

tab = [-3, 4.5, 6, 6, 3,3, 4,-4]

print(ordre_croissant(tab))

#tab = [5, 4, 6, 6, 3, 3, 4]
```

9. Créez une fonction qui prend un tableau NumPy de nombres entiers en entrée et renvoie un nouveau tableau contenant uniquement les nombres premiers du tableau d'origine.

```python
# 2️⃣ Fonction pour tester si un nombre est premier
def isPrime(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True    

def est_premier(tab):
    premier = []
    for i in tab :
        if isPrime(i) :
            premier.append(int(i))
    return premier

f = np.random.randint(20,size=(10)) 
print(f)
print(est_premier(f))
```

2. Boucle de jeu : 

```python
#! boucle de tentative
#! boucle de if in mot
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import numpy as np

def afficher_pendu(nb_erreurs):
    """
    Affiche le pendu en fonction du nombre d'erreurs.
    """

    plt.figure()  # Crée une nouvelle figure

    # Dessine la potence
    plt.plot([0, 1], [0, 0], 'k')  # Base
    plt.plot([0.5, 0.5], [0, 1], 'k')  # Poteau vertical
    plt.plot([0.5, 1], [1, 1], 'k')  # Barre horizontale
    plt.plot([1, 1], [0.9, 0.8], 'k')  # Corde

    # Dessine le corps du pendu en fonction du nombre d'erreurs
    if nb_erreurs < 6:# Tête
        cercle = plt.Circle((1, 0.75), 0.05, fill=False, color='k')
        plt.gca().add_patch(cercle)

    if nb_erreurs < 5 :# Corps
        plt.plot([1,1],[0.7,0.5],color='k')
    if nb_erreurs < 4:# Bras gauche
        plt.plot([1,0.9],[0.7,0.6],color='k')
    if nb_erreurs < 3:# Bras droit
        plt.plot([1,1.1],[0.7,0.6],color='k')
    if nb_erreurs <2:# Jambe gauche
        plt.plot([1,1.1],[0.5,0.3],color='k')
    if nb_erreurs <1 :# Jambe droite
        plt.plot([1,0.9],[0.5,0.3],color='k')
    # Paramètres d'affichage²
    plt.axis('on')  # Cache les axes
    plt.xlim(-0.2, 1.4)  # Ajuste les limites de l'axe x
    plt.ylim(-0.2, 1.2)  # Ajuste les limites de l'axe y
    plt.show()  # Affiche le pendu

nb_erreurs = 6
nouveauMot = list(MotMasque) 

while nb_erreurs > 0 :
    lettre = input("tu choisis quoi comme lettre")
    if lettre in ChoixMot : 
        index = ChoixMot.index(lettre)
        nouveauMot[index] = lettre
        print("Bravo tu as trouvé une lettre continue", *nouveauMot)
        
    else :
        nb_erreurs = nb_erreurs-1
        print("Raté ! Il te reste", nb_erreurs, "erreurs.")
        print(*nouveauMot)
        afficher_pendu(nb_erreurs)
    
    if "_" not in nouveauMot:
        print("Gagné !!")
        break
    if nb_erreurs == 0 : 
        print("Tu as perdu, Voici le mot : ", ChoixMot)



```
